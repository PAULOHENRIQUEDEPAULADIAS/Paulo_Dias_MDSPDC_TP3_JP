1. Dados como serviço, ou serviço de dados (DaaS), é o propósito de fornecer dados como um produto final. Ou seja, o dado é coletado bruto de uma fonte específica e passa por um processo de tratamento e robustez, no qual é processado para que se torne adequado ao uso. A partir disso, ele pode ser disponibilizado de forma confiável e fidedigna para o propósito necessário. Por exemplo, a API do Twitter fornece um DaaS: nela, os tweets são estruturados e separados em metadados, informações dos posts, autor, região e diversas outras características. Tudo isso é processado para que, posteriormente, alguém possa se conectar à API do Twitter e utilizar esses dados para pesquisas, análises, estudos de opinião, investigações de mercado, entre outros fins.

2. As tecnologias consideradas devem ser aquelas que oferecem suporte direto às características essenciais do ambiente de computação em nuvem. Por exemplo, a nuvem é um ambiente elástico, portanto a tecnologia escolhida também deve ser capaz de escalar de forma elástica conforme a demanda. Da mesma forma, serviços em nuvem oferecem fortes mecanismos de segurança, como criptografia e controle avançado de acesso, e a tecnologia utilizada precisa ser compatível com esses padrões. Outras propriedades importantes incluem tolerância a falhas, já que sistemas cloud-native operam em ambientes distribuídos e precisam continuar funcionando mesmo diante de falhas pontuais. Também é necessário considerar desempenho e baixa latência, pois a nuvem proporciona acesso rápido e eficiente aos dados, e a tecnologia adotada deve ser capaz de acompanhar esse nível de performance. O custo-benefício é outro aspecto relevante. Como os serviços em nuvem geralmente possuem modelos de cobrança flexíveis e econômicos, a tecnologia escolhida deve agregar nesse sentido, evitando desperdício de recursos e se ajustando bem ao modelo de escalabilidade e cobrança utilizado pela nuvem. Finalmente, é essencial que a tecnologia seja compatível com a infraestrutura cloud, integrando-se bem aos serviços nativos e às ferramentas do provedor escolhido.

3. O Flyway é uma ferramenta de gerenciamento de banco de dados voltada principalmente para o controle de versionamento. Ele cria automaticamente uma tabela interna de controle, onde registra todas as alterações aplicadas — como as queries executadas, o usuário responsável, a data e a hora — garantindo rastreabilidade e organização do histórico. Por ser baseado em scripts versionados, o Flyway facilita a reprodutibilidade dos esquemas e dados iniciais, sendo muito utilizado em ambientes de teste e desenvolvimento. Sua finalidade é padronizar e aplicar as mudanças no banco de forma consistente, evitando divergências entre ambientes e assegurando que todas as atualizações sigam a mesma sequência e estrutura. Isso é especialmente importante em sistemas que lidam com grandes volumes de dados e exigem alta confiabilidade.

4. A programação reativa é um modelo baseado em eventos. Nela, o sistema reage às ações que ocorrem, como entrada de dados, cliques, atualizações ou mensagens, e executa operações em resposta, sem depender de uma sequência rígida e síncrona de etapas. Diferente do modelo tradicional, onde existe um fluxo pré-definido (passo 1, passo 2, passo 3), a programação reativa trabalha de forma assíncrona, permitindo que cada evento seja tratado no momento em que acontece. Esse paradigma é construído sobre quatro características principais:
* Ele é responsivo, respondendo rapidamente aos eventos e mantendo uma boa experiência ao usuário.
* É resiliente, preparado para lidar com falhas sem comprometer o funcionamento do sistema.
* É elástico, capaz de adaptar-se automaticamente à demanda, escalando para mais ou menos recursos conforme necessário.
* E é baseado em mensagens, utilizando comunicação assíncrona e não bloqueante entre componentes, o que garante autonomia e eficiência.

5. A programação reativa é especialmente importante para aplicativos nativos em nuvem porque seus quatro pilares, responsividade, resiliência, elasticidade e comunicação orientada a mensagens, se alinham perfeitamente às necessidades desse tipo de arquitetura. Aplicações em nuvem precisam se adaptar rapidamente à variação de carga, lidar com falhas de forma isolada, responder aos usuários em tempo real e comunicar componentes distribuídos de maneira assíncrona. O modelo reativo atende a todas essas exigências ao trabalhar com fluxos de dados independentes, assíncronos e não bloqueantes, tornando o sistema mais eficiente, escalável e preparado para ambientes distribuídos. Em contraste, a programação imperativa segue um modelo síncrono e sequencial, onde cada operação depende da conclusão da anterior (passo 1 → passo 2 → passo 3). É uma abordagem mais rígida, que tende a bloquear processamento e a limitar a escalabilidade, especialmente em sistemas distribuídos. Já a programação reativa, por ser assíncrona, orientada a eventos e altamente adaptável, se torna muito mais adequada para aplicativos modernos em nuvem, que exigem flexibilidade, paralelismo e alta disponibilidade.

6. O Spring WebFlux é um módulo do Spring Framework criado para permitir o desenvolvimento de aplicações web reativas, assíncronas e não bloqueantes no lado do servidor. Ele foi introduzido a partir do Spring 5 e utiliza o Reactor, a biblioteca de programação reativa da JVM, que trabalha com os tipos Mono (0 a 1 elemento) e Flux (0 a N elementos). Esse módulo possibilita que aplicações Spring adotem uma arquitetura totalmente reativa, capaz de lidar com alta concorrência e fluxos de dados assíncronos de maneira eficiente. Dentro do WebFlux, existem duas maneiras principais de definir endpoints RESTful. A primeira é o uso do modelo tradicional baseado em anotações, semelhante ao Spring MVC. Nessa abordagem, criam-se classes anotadas com @RestController, e cada método é associado aos endpoints através de anotações como @GetMapping ou @PostMapping. Esses métodos retornam tipos reativos, garantindo o processamento não bloqueante compatível com o WebFlux. A segunda opção é o modelo funcional, que dispensa anotações e utiliza funções puras para configurar rotas e tratar requisições. Nele, são definidos dois componentes: o Handler, que contém os métodos responsáveis por processar as requisições e retornar um ServerResponse, e o Router, que especifica as rotas utilizando a API funcional do WebFlux, associando cada caminho HTTP ao método correspondente do Handler. As rotas são normalmente registradas como beans do Spring. Essa abordagem reflete com mais fidelidade os princípios da programação reativa, permitindo maior flexibilidade e composição funcional.

### Parte 3 Implementação de clientes reativos com WebClient

* Exercício 7: 

O Spring Framework vem com dois clientes que executam requisições HTTP. Quais são eles e como eles funcionam?

* Exercício 8: 

Quais as principais diferenças entre o RestTemplate e o WebClient?

7. Os dois clientes utilizados pelo Spring Framework para realizar requisições HTTP são o RestTemplate, do ecossistema Spring MVC, e o WebClient, do módulo Spring WebFlux. O RestTemplate segue o modelo tradicional de execução síncrona e bloqueante, típico da arquitetura imperativa. Isso significa que, ao fazer uma requisição, a thread fica parada aguardando a resposta para só então continuar o fluxo, funcionando de maneira sequencial, como uma fila de execução. Já o WebClient opera de forma assíncrona e não bloqueante, alinhado ao paradigma reativo do Spring WebFlux. Nesse modelo, as operações podem ocorrer em paralelo, pois são disparadas e tratadas por meio de eventos, sem necessidade de bloquear threads enquanto aguardam respostas. Isso permite maior escalabilidade e melhor desempenho em aplicações modernas.

8. As principais diferenças entre o RestTemplate e o WebClient estão no modelo de execução e no paradigma que cada um segue. O RestTemplate pertence ao ecossistema Spring MVC e funciona de forma síncrona e bloqueante, típica da arquitetura imperativa. Nesse modelo, cada requisição ocupa uma thread e precisa aguardar toda a cadeia de processamento para retornar, o que torna as chamadas sequenciais, funcionando praticamente como uma fila: uma requisição inicia, bloqueia a thread até receber a resposta e só então o fluxo continua. Já o WebClient opera de maneira assíncrona e não bloqueante, sendo parte do módulo Spring WebFlux e alinhado ao paradigma de programação reativa. Aqui, as operações são disparadas e tratadas por meio de eventos, permitindo que diversos fluxos ocorram em paralelo sem bloquear threads. Isso elimina a necessidade de uma fila de execução tradicional, reduz delays e permite lidar com alto volume de requisições de forma mais escalável e eficiente.

9. O Testcontainers é uma biblioteca que permite rodar, durante os testes, os mesmos tipos de serviços reais que sua aplicação usa em produção, por exemplo, um banco PostgreSQL de verdade, um Redis de verdade etc. Ele faz isso criando containers Docker temporários.Então, ao invés de você usar mocks ou versões “de mentirinha”, você testa sua aplicação conversando com um serviço real de verdade, só que isolado e criado só para os testes. Isso deixa os testes muito parecidos com o que acontece no ambiente de produção, mas sem usar o ambiente de produção de verdade — apenas cópias realistas feitas com imagens de Docker.

10. Testcontainers é recomendado quando é necessário testar aplicações que interagem com serviços externos, como bancos de dados, ferramentas de mensageria ou serviços de armazenamento. Ele cria essas dependências de forma isolada e automática usando containers Docker, dispensando configurações manuais e garantindo que cada teste execute em um ambiente limpo, confiável e idêntico ao anterior. Embora seja muito utilizado em cenários de aplicações nativas em nuvem, seu uso não se limita a eles. O Testcontainers pode ser adotado em qualquer tipo de aplicação que dependa desses serviços externos, oferecendo um ambiente de teste realista ao iniciar as imagens apenas durante a execução dos testes e destruí-las logo em seguida.
